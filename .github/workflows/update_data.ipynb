```python
# Nommez ce fichier : update_data.py
import yfinance as yf
import pandas as pd
from datetime import datetime

# --- Paramètres ---
ticker_symbol = "AAPL"  # Ou "MSFT", "GOOG", etc.
start_date = "2020-01-01"
end_date = datetime.now().strftime("%Y-%m-%d") # Prend la date du jour

print(f"Lancement de la mise à jour pour {ticker_symbol}...")

try:
    data = yf.download(ticker_symbol, start=start_date, end=end_date)
    
    if not data.empty:
        data_reset = data.reset_index().rename(columns={'Date': 'date'})
        columns_to_keep = ['date', 'Open', 'High', 'Low', 'Close', 'Volume']
        df_final = data_reset[columns_to_keep]
        
        file_name = f"{ticker_symbol}_historique_bourse.csv"
        df_final.to_csv(file_name, index=False)
        
        print(f"Succès : Fichier {file_name} mis à jour avec les données jusqu'au {end_date}.")
        print("Dernières 5 lignes :")
        print(df_final.tail())
    else:
        print("Aucune donnée téléchargée.")

except Exception as e:
    print(f"Erreur lors de l'exécution du script : {e}")
```
